# 1. Creation of a dictionary with lyrics

```python
import requests
from bs4 import Tag, NavigableString, BeautifulSoup
# import re
import pandas as pd
import random
import time
```

```python

artist = "depeche_mode"

# done:

# "imagine_dragons"
# "rainbow"
# "u2"
# "roxette"
# "kasabian"  # a little
# "blondie"
# "doors"
# "guns_n_roses"
# "lana_del_rey"  # a little
# "florence_the_machine"
# "franz_ferdinand"
# "bob_dylan"
# "queen"
# "muse"
# "florence_the_machine"
```

```python

url = 'https://www.allthelyrics.com/lyrics/' + artist
req_announcement = requests.get(url)
html_doc = req_announcement.text
# print(html_doc)

soup = BeautifulSoup(html_doc, 'html.parser')
kek = soup.find_all('div', attrs={'id': ['lyricslist-left', 'lyricslist-right']})
# kek
```

```python
discography = {}

for k in kek:
    uuu = k.find_all('li')
    for u in uuu:
        discography[str.strip(u.text)] = {
            'link': "https://www.allthelyrics.com" + u.find('a')['href']
        }
        print(u.text)
```

```python
# discography
```

```python
# I need to make pauses because of blocking
counter = 0

for song in discography.keys():
    counter += 1
    print(counter, song)
    time.sleep(1)
    if counter % 10 == 0:
        time.sleep(4)

    req_announcement2 = requests.get(discography[song]['link'])
    html_doc2 = req_announcement2.text
    soup2 = BeautifulSoup(html_doc2, 'html.parser')
    
    kek2 = soup2.find_all('div', attrs={'class': ['content-text-album']})
    if kek2 == []:
        discography[song]['album'] = '-'
    else:
        discography[song]['album'] = str.strip(kek2[0].text)
    
    kek3 = soup2.find_all('div', attrs={'class': ['content-text-inner']})
    if kek3 == []:
        discography[song]['lyrics'] = '-'
    else:
        discography[song]['lyrics'] = str.strip(kek3[0].text)
```

```python
# discography
```

```python
discography_df = pd.DataFrame.from_dict({(i): discography[i] 
                           for i in discography.keys()},
                       orient='index')
# discography_df

discography_df_2 = discography_df.reset_index().rename(columns={'index':'song'})
discography_df_2
```

```python

def add_gaps_to_lyrics(discography_df, song_title, string_frequency):
    
    song_txt = discography_df[discography_df['song'] == song_title]['lyrics'].iloc[0]
    # print(song_txt)
    song_strings = song_txt.split('\n')
    # song_strings

    string_counter = 0
    correct_answers = []
    gap_counter = 0
    
    fixed_song_list = []

    for s in song_strings:
        if string_counter % string_frequency == 0:
            gap_counter += 1
            words = s.split(' ')
            deleted_word_no = random.randint(0,len(words)-1)
            correct_answers.append(str.lower(words[deleted_word_no]))
            words[deleted_word_no] = "({0})".format(gap_counter) + '_'*len(words[deleted_word_no])
            fixed_song_list.append(' '.join(words))
#             print(' '.join(words))
        else:
            fixed_song_list.append(s)
#             print(s)
        string_counter += 1
    
    with open("LYRICS_WITH_GAPS.txt",'w') as file:
        pass
    with open('LYRICS_WITH_GAPS.txt', 'w') as f:
        for s in fixed_song_list:
            f.write(f"{s}\n")
    
    return correct_answers, gap_counter
```

```python

def mark(a, b):
    if a == b:
        return '+'
    else:
        return ''

def answers_processing(correct_answers, gap_counter):

    my_answers = []

    for i in range(gap_counter):
#         print("{0}:".format(i+1))
        my_answers.append(str.lower(str.strip(input("{0}:\t ".format(i+1)))))

    print()
    correct_cnt = sum([1 for i in range(gap_counter) if correct_answers[i] == my_answers[i] ])
#     print("Results: {0} from {1} ({2}%)".format(correct_cnt, gap_counter, round(correct_cnt/gap_counter, 1)))

    shift = max(8, max(map(len, correct_answers)))
    print("\t correct:{0} - my answer:".format(' '*(shift-8)))

    for i in range(gap_counter):
        print("{0}  {1}\t {2} - {3}".format(i+1, mark(correct_answers[i],my_answers[i]), correct_answers[i] + ' '*(shift - len(correct_answers[i])), my_answers[i]))
    print()

    return "Results: {0} from {1} ({2}%)".format(correct_cnt, gap_counter, round(100 * correct_cnt/gap_counter, 1))
```

```python
'''
Choose the song:
'''

song_title = 'No Light No Light'
string_frequency = 2
```

```python
song_txt = discography_df[discography_df['song'] == song_title]['lyrics'].iloc[0]
print(song_txt)
```

```python
'''
Lyrics with gaps
'''

correct_answers, gap_counter = add_gaps_to_lyrics(discography_df, song_title, string_frequency)
```

```python
'''
Fill the gaps
'''

print(answers_processing(correct_answers, gap_counter))
```
