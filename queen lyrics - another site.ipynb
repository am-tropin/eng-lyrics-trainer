# 1. Creation of a dictionary with links to all Queen sonds

```python
import requests
from bs4 import Tag, NavigableString, BeautifulSoup
import re

import pandas as pd

import random
```

```python

queen_url = 'https://www.allthelyrics.com/lyrics/queen'
req_announcement = requests.get(queen_url)
html_doc = req_announcement.text
print(html_doc)
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
kek = soup.find_all('div', attrs={'id': ['lyricslist-left', 'lyricslist-right']})
kek
```

```python
queen_discography = {}

for k in kek:
    uuu = k.find_all('li')
    for u in uuu:
        queen_discography[str.strip(u.text)] = {
            'link': "https://www.allthelyrics.com" + u.find('a')['href']
        }
        print(u.text)
```

```python
queen_discography
```

```python
import time
```

```python
# I need to make pauses because of blocking
counter = 0

for song in queen_discography.keys():
    counter += 1
    print(counter, song)
    time.sleep(1)
    if counter % 10 == 0:
        time.sleep(9)

    req_announcement2 = requests.get(queen_discography[song]['link'])
    html_doc2 = req_announcement2.text
    soup2 = BeautifulSoup(html_doc2, 'html.parser')
    
    kek2 = soup2.find_all('div', attrs={'class': ['content-text-album']})
    if kek2 == []:
        queen_discography[song]['album'] = '-'
    else:
        queen_discography[song]['album'] = str.strip(kek2[0].text)
    
    kek3 = soup2.find_all('div', attrs={'class': ['content-text-inner']})
    if kek3 == []:
        queen_discography[song]['lyrics'] = '-'
    else:
        queen_discography[song]['lyrics'] = str.strip(kek3[0].text)
```

```python
queen_discography
```

```python
queen_discography_df = pd.DataFrame.from_dict({(i): queen_discography[i] 
                           for i in queen_discography.keys()},
                       orient='index')
queen_discography_df
```

```python
queen_discography_df_2 = queen_discography_df.reset_index().rename(columns={'index':'song'})
queen_discography_df_2
```

```python
queen_discography_df_2.to_csv('queen_discography_allthelyrics.csv', sep='\t', encoding='utf-8', index=False)
```
