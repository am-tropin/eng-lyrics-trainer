# 1. Creation of a dictionary with links to all Queen sonds

```python
import requests
from bs4 import Tag, NavigableString, BeautifulSoup
import re

import pandas as pd

import random
```

```python
queen_discography = {}

queen_url = 'https://www.azlyrics.com/q/queen.html'
req_announcement = requests.get(queen_url)
html_doc = req_announcement.text
# print(html_doc)

soup = BeautifulSoup(html_doc, 'html.parser')
kek = soup.find_all('div', attrs={'class': ['album', 'listalbum-item']})
    
for k in kek:
    if k['class'][0] == 'album':
        temp_album = k.text
    elif k['class'][0] == 'listalbum-item':
        queen_discography[k.text] = {
            'album': temp_album[7:]
        }
        if k.find('a'):
            queen_discography[k.text]['link'] = k.find('a')['href']
        else:
            queen_discography[k.text]['link'] = '-'
        
queen_discography
```

```python
queen_discography_df = pd.DataFrame.from_dict({(i): queen_discography[i] 
                           for i in queen_discography.keys()},
                       orient='index')
queen_discography_df
```

```python
queen_discography_df_2 = queen_discography_df.reset_index().rename(columns={'index':'song'})
queen_discography_df_2
```

```python
queen_discography_df_2.to_csv('queen_discography_links.csv', sep='\t', encoding='utf-8', index=False)
```

```python
# I need a timer because of blocking

for song_title in queen_discography.keys():
    if queen_discography[song_title]['link'] == '-':
        queen_discography[song_title]['lyrics'] = ''
    else:
        print(song_title)
        req_announcement2 = requests.get('https://www.azlyrics.com' + queen_discography[song_title]['link'])
        html_doc2 = req_announcement2.text
        soup2 = BeautifulSoup(html_doc2, 'html.parser')
        kek2 = soup2.find_all('div', attrs={'class': ['container main-page']})
        kek22 = kek2[0].find_all('div', attrs={'class': ['row']})
        kek23 = kek22[0].find_all('div', attrs={'class': ['col-xs-12 col-lg-8 text-center']})
        kek24 = kek23[0].find_all('div', attrs={'class': None})
        queen_discography[song_title]['lyrics'] = str.strip(kek24[0].text)
```
