# 1. Creation of a dictionary with links to all Queen sonds

```python
queen_url = 'https://www.azlyrics.com/q/queen.html'
```

```python
import requests
from bs4 import Tag, NavigableString, BeautifulSoup
import re

import pandas as pd

import random
```

```python
req_announcement = requests.get(queen_url)
html_doc = req_announcement.text
print(html_doc)
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')

listAlbum = soup.find('div', attrs={'id': 'listAlbum'})
print(listAlbum)
```

```python
album = soup.find_all('div', attrs={'class': 'album'})

for a in album:
    print(a)
```

```python
listalbum = soup.find_all('div', attrs={'class': 'listalbum-item'})

for a in listalbum:
    print(a)
```

```python
kek = soup.find_all('div', attrs={'class': ['album', 'listalbum-item']})
    
for k in kek:
    print(k)
#     if k.find('div', attrs={'class': 'album'}):
#         print('AAAAAA!')
    print(k.find('div', attrs={'class': 'album'}))
```

```python
kek = soup.find_all('div')

for k in kek:
    if k['div'] == 'album':
        print(k)
```
